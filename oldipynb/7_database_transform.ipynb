```python
import cobra
import re
#model = cobra.io.read_sbml_model(org_name+'_model\\'+org_name+'_add_from_modelseed_6.xml')
metamapping={}
errorlist = {}
for meta in model.metabolites:
    bigg_meta = meta.annotation.get('bigg.metabolite')
    if type(bigg_meta) == list:
        bigg_meta = bigg_meta[0]
    if bigg_meta:
        match = re.search(r'cpd\d+', meta.id)
        if match:
            metaID_string = meta.id.replace(match.group(),bigg_meta)
            metamapping[meta.id] = metaID_string
            #print(meta.id+' replaced by '+metaID_string)
            try:
                meta.id = metaID_string
            except:
                errorlist[meta.id] = metaID_string
                pass

for rxn in model.reactions:
    bigg_rxn = rxn.annotation.get('bigg.reaction')
    if type(bigg_rxn) == list:
        bigg_rxn = bigg_rxn[0]
    if bigg_rxn:
        match = re.search(r'rxn\d+', rxn.id)
        if match:
            rxnID_string = rxn.id.replace(match.group(),bigg_rxn)
            #print(rxn.id+' replaced by '+rxnID_string)
            try:
                rxn.id = rxnID_string
            except:
                errorlist[meta.id] = rxnID_string
                pass
    if 'cpd' in rxn.id:
        match = re.search(r'cpd\w*', rxn.id)
        if match.group() in metamapping.keys():
            rxnID_string = rxn.id.replace(match.group(),metamapping.get(match.group()))
            rxn.id = rxnID_string

for key, value in errorlist.items():
    if model.metabolites.get_by_id(value):
        for rxn in model.reactions:
            if rxn.metabolites.get(key):
                rxn.add_metabolites({value:rxn.metabolites.get(key)})
                rxn.remove_metabolites({key:rxn.metabolites.get(key)})
                model.remove_metabolites(model.metabolites.get_by_id(key))


cobra.io.write_sbml_model(model,filename=org_name+'_model\\'+org_name+'_add_from_modelseed_to_BIGG.xml')
model
```
