```python
org_name = 'mko'
org_full_name = 'Methylomonas_koyamae'
```

```python
import cobra

model = cobra.io.read_sbml_model(org_name+'_model\\'+org_name+'_bigg_model.xml')
carveme_model = cobra.io.read_sbml_model(org_name+'_model\\'+org_name+'_carveme_model.xml')
```

```python
iMcBath_model = cobra.io.read_sbml_model('iMcBath.xml')
iMcBath_model
```

```python
modellist=[r.id for r in model.reactions]
carvemelist=[r.id for r in carveme_model.reactions]
iMcBathlist=[r.id for r in iMcBath_model.reactions]
```

```python
modeliniMcBath = [r for r in modellist if r in iMcBathlist]
modelnotiniMcBath = [r for r in modellist if r not in iMcBathlist]
iMcBathinmodel = [r for r in iMcBathlist if r in modellist]
iMcBathnotinmodel = [r for r in iMcBathlist if r not in modellist]
print('modeliniMcBath:',len(modeliniMcBath))
print('modelnotiniMcBath:',len(modelnotiniMcBath))
print('iMcBathinmodel:',len(iMcBathinmodel))
print('iMcBathnotinmodel:',len(iMcBathnotinmodel))
```

```python
iMcBathnotinmodelincarveme = [r for r in iMcBathnotinmodel if r in carvemelist]
print('iMcBathnotinmodelincarveme:',len(iMcBathnotinmodelincarveme))
```

```python
for r in modeliniMcBath:
    print(r)
    print('     '+model.reactions.get_by_id(r).name)
```
