```python
fr=open('Database//keggrxn_fromRAVEN.txt','r',encoding='utf-8-sig')
with open('Database//keggrxn_fromRAVENnew.txt','w',encoding='utf-8-sig') as fw:
    for line in fr:
        data = line.split('\t')
        if ";" in data[4]:
            subsystems = data[4].split(';')
            rnid=""
            for subsystem in subsystems:
                try:
                    searchid = pathwaymap_df[pathwaymap_df[0]==subsystem][1].values[0]
                except:
                    searchid = ""
                rnid=rnid+searchid+";"
        else:
            try:
                rnid = pathwaymap_df[pathwaymap_df[0]==data[4]][1].values[0]
            except:
                rnid= ""   
        fw.write(data[0]+'\t'+data[1]+'\t'+data[2]+'\t'+data[3]+'\t'+data[4]+'\t'+rnid+'\n')
fw.close()
```

```python
from PIL import Image, ImageFont, ImageDraw

img = Image.open("map00680.png")

# 创建一个ImageDraw对象
draw = ImageDraw.Draw(img)

# 选择一个字体文件和字体大小
font = ImageFont.truetype("arial.TTF", 18)

# 计算文本的宽度和高度
box = draw.textbbox((0,0),"TEXT", font=font)
w = box[2] - box[0]
h = box[3] - box[1]


# 设置文本的位置和颜色
pos = (278-w/2, 279-h/2)
color = (0, 0, 0)

# 在图片上绘制矩形
draw.rectangle((232, 262, 324, 296), fill=(255, 192, 203))

# 在图片上绘制文本
draw.text(pos, "TEXT", fill=color, font=font)

# 保存修改后的图片
img.save("SAVE.png")
```
